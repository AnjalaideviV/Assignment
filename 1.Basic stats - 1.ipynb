# 3.Standardization of Numerical Variables

```python
scaler_values=scaler.fit_transform(df[['Avg Price','Net Sales Value']])
scaler_values
```

```python
# the distribution of data after standardization
plt.scatter(x=scaler_values[:,0], y=scaler_values[:,1])
```
